```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zad2a = [np.random.uniform(low = 0.0, high = 1.0, size = None) for i in range(1000)]
zad2b = np.random.normal(0,1,1000)
zad2c = np.random.normal(0,10,1000)
zad2da = np.random.beta(1,1,1000)
zad2db = np.random.beta(2,2,1000)
zad2dc = np.random.beta(3,9,1000)
zad2ea = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10000)
zad2eb = np.random.multivariate_normal([0, 0], [[2, 1], [1, 2]], 10000)
```

```python
def plot_1Dhistogram(x, bin=10):
  plt.figure()
  plt.hist(x,bins=bin, density=True)
  plt.xlabel('values')
  plt.ylabel('normalised number of samples')
  plt.grid()
  plt.show()
  return

plot_1Dhistogram(zad2a,50)
plot_1Dhistogram(zad2b,15)
plot_1Dhistogram(zad2c,15)
plot_1Dhistogram(zad2da,50)
plot_1Dhistogram(zad2db,15)
plot_1Dhistogram(zad2dc,20)
```

```python
def plot_2Dhistogram_byplt(x, bin=50):
  plt.figure()
  plt.hist2d(x[:,0],x[:,1], bins=bin, density=True)
  plt.xlabel('x values')
  plt.ylabel('y_values')
  plt.grid()
  plt.show()
  return

def plot_2Dhistogram_bysbn(x, bin=50):
  sns.histplot(x=x[:,0],y=x[:,1])
  return

plot_2Dhistogram_byplt(zad2ea)
plot_2Dhistogram_bysbn(zad2ea)
plot_2Dhistogram_byplt(zad2eb)
plot_2Dhistogram_bysbn(zad2eb)
```

```python
alpha = np.random.normal(2,0.1,1000)
beta = np.random.normal(0.2,0.2,1000)
x = np.random.normal(0,3,1000)
sigma = np.abs(np.random.normal(0,1,1000))
mi = alpha + (beta * x)
y = np.random.normal(mi,sigma,1000)
```

```python
x1 = np.linspace(-6,6,70)
plt.figure()
plt.plot(x,y, '.')
for i in range(70):
  y1 = alpha[i] + (beta[i] * x1)
  plt.plot(x1, y1, 'r')
plt.xlabel('x values')
plt.ylabel('y_values')
plt.grid()
plt.show()
```
